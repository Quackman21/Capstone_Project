```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Specify the file path
file_path = '/Users/ivanquackenbush/Documents/44-608/Capstone/Capstone_Project/Clean Training Data.csv'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few entries of the DataFrame
(df.head())
```

```python
# Group data by player name and calculate average for each player
grouped_df = df.groupby('Player Name').mean()

# Calculate standard deviations
std_batspeed = grouped_df['Bat Speed (mph)'].std()
std_rotational_acceleration = grouped_df['Rotational Acceleration (g)'].std()
std_on_plane_efficiency = grouped_df['On Plane Efficiency (%)'].std()

# Calculate composite score for each player
weight_batspeed = 1.75
weight_rotational_acceleration = 1.5
weight_on_plane_efficiency = 1.0
std_weight = 1/5  # Standard deviation weight (1/3 of overall composite score)

# Scale down standard deviations weight
scaled_std_batspeed = std_batspeed / 3.1
scaled_std_rotational_acceleration = std_rotational_acceleration / 3.1
scaled_std_on_plane_efficiency = std_on_plane_efficiency / 3

grouped_df['Composite Score'] = (
    (weight_batspeed * grouped_df['Bat Speed (mph)']) +
    (weight_rotational_acceleration * grouped_df['Rotational Acceleration (g)']) +
    (weight_on_plane_efficiency * grouped_df['On Plane Efficiency (%)']) +
    (std_weight * (scaled_std_batspeed + scaled_std_rotational_acceleration + scaled_std_on_plane_efficiency))
)

# Rank players based on composite score
grouped_df['Rank'] = grouped_df['Composite Score'].rank(ascending=False)

# Print the names along with their composite score and rank
print(grouped_df[['Composite Score', 'Rank']].sort_values(by='Rank'))
```
