```python
#Create a graph getting a count of each players swings

import matplotlib.pyplot as plt

swing_counts = df['Player Name'].value_counts()

# Create a bar plot
ax = swing_counts.plot(kind='bar', figsize=(10, 6))

# Set the title and labels
plt.title('Number of Swings Per Player')
plt.xlabel('Player Name')
plt.ylabel('Number of Swings')

# Rotate the player names 
plt.xticks(rotation=45)

# Adding data labels
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

# Show the plot
plt.show()
```

```python
#Produce a Scatter plot for the entire teams Swings

import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(12, 6))
for name, group in df.groupby('Player Name'):
    plt.scatter(group['Rotational Acceleration (g)'], group['Bat Speed (mph)'], label=name)

# Set the title and labels
plt.title('Rotational Acceleration vs. Bat Speed by Player')
plt.xlabel('Rotational Acceleration (g)')
plt.ylabel('Bat Speed (mph)')
plt.legend(title='Player Name')

# Plot
plt.show()
```

```python
#Individual player rotational acceleration vs batspeed

import matplotlib.pyplot as plt

# Get unique player names
players = df['Player Name'].unique()

# Create a figure and an array of subplots with 1 column and as many rows as there are players
fig, axes = plt.subplots(len(players), 1, figsize=(4, 4 * len(players)))  # 4x4 for each subplot

# Check if axes is not an array (happens when there's only one subplot)
if len(players) == 1:
    axes = [axes]

# Loop through each player and plot on its own subplot
for ax, player in zip(axes, players):
    # Filter the dataset for the current player
    player_data = df[df['Player Name'] == player]
    
    # Scatter plot for each player on its own subplot axis
    ax.scatter(player_data['Rotational Acceleration (g)'], player_data['Bat Speed (mph)'])
    
    # Set title and labels for each subplot
    ax.set_title(player)
    ax.set_xlabel('Rotational Acceleration (g)')
    ax.set_ylabel('Bat Speed (mph)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Create distrubution graphs for one plane efficiency 

import seaborn as sns
import matplotlib.pyplot as plt

# Get unique player names
players = df['Player Name'].unique()

# Create a figure and an array of subplots with 1 column and as many rows as there are players
fig, axes = plt.subplots(len(players), 1, figsize=(4, 4 * len(players)))  # Adjust the figure size as needed

# Ensure axes is iterable in case of a single subplot scenario
if len(players) == 1:
    axes = [axes]

# Loop through each player and plot on its own subplot
for ax, player in zip(axes, players):
    # Filter the dataset for the current player
    player_data = df[df['Player Name'] == player]
    
    # Use seaborn's distplot to plot a histogram with a KDE overlay for each player
    sns.histplot(player_data['On Plane Efficiency (%)'], bins=20, kde=True, ax=ax,
                 stat="density", linewidth=0, alpha=0.5, color='blue')
    
    # Set title and labels for each subplot
    ax.set_title(f'On Plane Efficiency (%) Distribution - {player}')
    ax.set_xlabel('On Plane Efficiency (%)')
    ax.set_ylabel('Density')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Load the Players Averages and Stanard Deviations

import pandas as pd
from IPython.display import display

# Group the DataFrame by 'Player Name' and calculate mean and std for the specified metrics
metrics = ['Bat Speed (mph)', 'Rotational Acceleration (g)', 'On Plane Efficiency (%)']
stats_df = df.groupby('Player Name')[metrics].agg(['mean', 'std']).reset_index()

# Flatten the MultiIndex in columns and create more readable column names
stats_df.columns = [' '.join(col).strip() for col in stats_df.columns.values]

# Rename columns for clarity
stats_df.columns = ['Player Name', 'Average Bat Speed (mph)', 'Std Bat Speed (mph)',
                    'Average Rotational Acceleration (g)', 'Std Rotational Acceleration (g)',
                    'Average On Plane Efficiency (%)', 'Std On Plane Efficiency (%)']

# Display the DataFrame as an HTML table
display(stats_df)
```
